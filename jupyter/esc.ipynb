```python
def preveri_pogoj(element):
    
    if type(element) == int:
        if (element >= 5) & (element <= 12):
            return True
        else:
            return False

    elif type(element) == str:
        b = element.split(',')
        integer_map = map(int, b)
        integer_list = list(integer_map)
        i=0
        for x in integer_list:
            if (x >= 5) & (x <= 12):
                integer_list[i] = True
                i = i + 1
            else:
                integer_list[i] = False
                i = i + 1

        for i in integer_list:
            if i == True:
                return True
                break
            
    elif type(element) == float:
        return False

maska = tabela['S'].apply(preveri_pogoj)
nova_tabela = tabela[maska]
#nova_tabela
```
